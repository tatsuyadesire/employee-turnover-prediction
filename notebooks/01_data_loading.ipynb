### Sprint 2 – Conhecimento dos Dados
-Construção de Dicionário de Dados
-Estatísticas Descritivas
-Análise da Variável Alvo

```python
#Dicionário - Coluna e Tipo de Dado

dicionario = pd.DataFrame({
    "Coluna": df.columns,
    "Tipo": df.dtypes.astype(str)
})

dicionario
```

```python
#Dicionario - Definir numéricas e categóricas

numericas = df.select_dtypes(
    include=["int64","float64"]
).columns

categoricas = df.select_dtypes(
    include=["object"]
).columns

print("Numéricas:", len(numericas))
print("Categóricas:", len(categoricas))
```

```python
#Dicionario - Coluna Categorias

dicionario["Categoria"
           ] = ""
dicionario.loc[
    dicionario["Coluna"] == "Attrition",
    "Categoria"
] = "Alvo"
dicionario.loc[
    dicionario["Coluna"].isin([
        "EmployeeNumber",
        "EmployeeCount"
    ]),
    "Categoria"
] = "Identificador"
dicionario.loc[
    dicionario["Tipo"] == "object",
    "Categoria"
] = "Categórica"

dicionario.loc[
    dicionario["Tipo"] != "object",
    "Categoria"
] = "Numérica"
```

```python
#Dicionario - Coluna Descrição

dicionario["Descricao"] = ""
descricoes = {
    "Age": "Idade do funcionário",
    "Attrition": "Indica se o funcionário deixou a empresa",
    "BusinessTravel": "Frequência de viagens corporativas",
    "DailyRate": "Taxa diária associada ao funcionário",
    "Department": "Departamento de atuação",
    "DistanceFromHome": "Distância entre residência e trabalho",
    "Education": "Nível de escolaridade",
    "EducationField": "Área de formação acadêmica",
    "EmployeeCount": "Quantidade de funcionários (constante)",
    "EmployeeNumber": "Identificador único do funcionário",
    "EnvironmentSatisfaction": "Satisfação com o ambiente de trabalho",
    "Gender": "Gênero do funcionário",
    "HourlyRate": "Taxa horária",
    "JobInvolvement": "Nível de envolvimento com o trabalho",
    "JobLevel": "Nível hierárquico",
    "JobRole": "Cargo ocupado",
    "JobSatisfaction": "Satisfação com o trabalho",
    "MaritalStatus": "Estado civil",
    "MonthlyIncome": "Renda mensal",
    "MonthlyRate": "Taxa mensal associada ao funcionário",
    "NumCompaniesWorked": "Quantidade de empresas onde trabalhou",
    "Over18": "Indica se possui mais de 18 anos",
    "OverTime": "Realiza horas extras",
    "PercentSalaryHike": "Percentual do último aumento salarial",
    "PerformanceRating": "Avaliação de desempenho",
    "RelationshipSatisfaction": "Satisfação nos relacionamentos profissionais",
    "StandardHours": "Horas padrão de trabalho",
    "StockOptionLevel": "Nível de benefício em ações",
    "TotalWorkingYears": "Anos totais de experiência",
    "TrainingTimesLastYear": "Quantidade de treinamentos no último ano",
    "WorkLifeBalance": "Equilíbrio entre vida pessoal e trabalho",
    "YearsAtCompany": "Tempo de empresa",
    "YearsInCurrentRole": "Tempo na função atual",
    "YearsSinceLastPromotion": "Tempo desde a última promoção",
    "YearsWithCurrManager": "Tempo com o gestor atual"
}

dicionario["Descricao"] = dicionario["Coluna"].map(descricoes)
```

```python
#Exportar dicionário

OUTPUT_DIR = ROOT_DIR / "outputs"

OUTPUT_DIR.mkdir(exist_ok=True)

dicionario.to_excel(
    OUTPUT_DIR / "data_dictionary.xlsx",
    index=False
)
```

```python
#Tabela Estatística

estatisticas = pd.DataFrame({
    "Media": df.mean(numeric_only=True),
    "Mediana": df.median(numeric_only=True),
    "Desvio_Padrao": df.std(numeric_only=True),
    "Minimo": df.min(numeric_only=True),
    "Maximo": df.max(numeric_only=True)
})

estatisticas
```

```python
#Arredondar estatísticas para 2 casas decimais

estatisticas = estatisticas.round(2)

estatisticas
```

```python
#Checar formato da tabela de estatísticas

estatisticas.shape
```

```python
#Exportar estatísticas descritivas

OUTPUT_FILE = ROOT_DIR / "outputs" / "descriptive_statistics.xlsx"

estatisticas.to_excel(
    OUTPUT_FILE
)
```

```python
#Quantidade absoluta de "Attriction"

attrition_count = df["Attrition"].value_counts()

print(attrition_count)
```

```python
#Percentual de "Attriction"

attrition_pct = (
    df["Attrition"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)

print(attrition_pct)
```

```python
#Gerar gráfico de barras para a distribuição de "Attrition"

attrition_count.plot(kind="bar")

plt.title("Distribuição de Attrition")
plt.xlabel("Attrition")
plt.ylabel("Quantidade de Funcionários")

plt.show()
```

#### Sprint 2 - Conslusão Analítica

Foi construído o dicionário de dados contendo a documentação das 35 variáveis presentes na base, incluindo classificação por tipo, categoria e descrição funcional.

A análise estatística descritiva permitiu compreender o comportamento das variáveis numéricas, identificando medidas centrais, dispersão e amplitude dos dados. Observou-se grande variabilidade em atributos relacionados à remuneração, experiência profissional e permanência na empresa, fatores potencialmente relevantes para o estudo de turnover.
