```python
#Importação de bibliotecas

import pandas as pd

from pathlib import Path

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
#Importação de datasets processados

ROOT_DIR = Path.cwd().parent

DATA_DIR = ROOT_DIR / "data" / "processed"

X_train = pd.read_csv(DATA_DIR / "X_train.csv")
X_test = pd.read_csv(DATA_DIR / "X_test.csv")

y_train = pd.read_csv(DATA_DIR / "y_train.csv")
y_test = pd.read_csv(DATA_DIR / "y_test.csv")
```

```python
# Conferir formatos

print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
# Ajustar formato de y_train e y_test

y_train = y_train.squeeze()
y_test = y_test.squeeze()
```

#### Modelo Baseline

```python
# Criar modelo baseline

modelo_lr = LogisticRegression(
    max_iter=1000,
    random_state=42
)
```

```python
# Treinar modelo

modelo_lr.fit(
    X_train,
    y_train
)
```

```python
# Fazer previsões usando a base de treino

y_pred_lr = modelo_lr.predict(X_test)
```

```python
# Métricas de avaliação

accuracy = accuracy_score(
    y_test,
    y_pred_lr
)

precision = precision_score(
    y_test,
    y_pred_lr
)

recall = recall_score(
    y_test,
    y_pred_lr
)

f1 = f1_score(
    y_test,
    y_pred_lr
)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)
```

```python
# Classification report

print(
    classification_report(
        y_test,
        y_pred_lr
    )
)
```

```python
# Matriz de confusão

print(
    confusion_matrix(
        y_test,
        y_pred_lr
    )
)
```

##### Conclusão - Modelo Baseline (Regressão Logística)

Foi treinado um modelo de Regressão Logística utilizando os dados preparados na Sprint 4.

O modelo atingiu Accuracy de 86,05%, Precision de 68,75%, Recall de 23,40% e F1 Score de 34,92%.

Apesar da boa Accuracy, observou-se baixa capacidade de identificação dos funcionários que efetivamente deixaram a empresa. O Recall de 23,40% indica que a maioria dos desligamentos não foi detectada pelo modelo.

Esse comportamento é esperado em bases desbalanceadas, nas quais a classe majoritária influencia fortemente o treinamento.

Os resultados obtidos serão utilizados como baseline para comparação com modelos mais robustos nas próximas etapas do projeto.

#### Modelo RandomForest

```python
# Importar biblioteca

from sklearn.ensemble import RandomForestClassifier
```

```python
# Criação do modelo Random Forest

modelo_rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
```

```python
# Treinar modelo

modelo_rf.fit(
    X_train,
    y_train
)
```

```python
# Fazer previsões usando a base de treino

y_pred_rf = modelo_rf.predict(
    X_test
)
```

```python
# Métricas de avaliação

accuracy_rf = accuracy_score(
    y_test,
    y_pred_rf
)

precision_rf = precision_score(
    y_test,
    y_pred_rf
)

recall_rf = recall_score(
    y_test,
    y_pred_rf
)

f1_rf = f1_score(
    y_test,
    y_pred_rf
)

print("Accuracy :", accuracy_rf)
print("Precision:", precision_rf)
print("Recall   :", recall_rf)
print("F1 Score :", f1_rf)
```

```python
# Classification report

print(
    classification_report(
        y_test,
        y_pred_rf
    )
)
```

```python
# Matriz de confusão

print(
    confusion_matrix(
        y_test,
        y_pred_rf
    )
)
```

##### Conclusão - Random Forest

O modelo Random Forest foi treinado utilizando os mesmos conjuntos de treino e teste empregados na Regressão Logística.

O modelo apresentou Accuracy de 83,33%, Precision de 41,67%, Recall de 10,64% e F1 Score de 16,95%.

Apesar de obter boa taxa de acerto global, o modelo demonstrou baixa capacidade de identificar funcionários que efetivamente deixaram a empresa, classificando corretamente apenas 5 dos 47 desligamentos observados na base de teste.

Comparado ao modelo baseline de Regressão Logística, o Random Forest apresentou desempenho inferior em todas as métricas relacionadas à identificação da classe de turnover.

#### Modelo XGBoost

```python
# Importar biblioteca

import xgboost
from xgboost import XGBClassifier
```

```python
# Criar modelo XGBoost

modelo_xgb = XGBClassifier(
    random_state=42,
    eval_metric="logloss"
)
```

```python
# Treinar modelo

modelo_xgb.fit(
    X_train,
    y_train
)
```

```python
# Fazer previsões usando a base de treino

y_pred_xgb = modelo_xgb.predict(
    X_test
)
```

```python
# Métricas de avaliação

accuracy_xgb = accuracy_score(
    y_test,
    y_pred_xgb
)

precision_xgb = precision_score(
    y_test,
    y_pred_xgb
)

recall_xgb = recall_score(
    y_test,
    y_pred_xgb
)

f1_xgb = f1_score(
    y_test,
    y_pred_xgb
)

print("Accuracy :", accuracy_xgb)
print("Precision:", precision_xgb)
print("Recall   :", recall_xgb)
print("F1 Score :", f1_xgb)
```

```python
# Classification report

print(
    classification_report(
        y_test,
        y_pred_xgb
    )
)
```

```python
# Matriz de confusão

print(
    confusion_matrix(
        y_test,
        y_pred_xgb
    )
)
```

```python
# RUCAUC

y_prob_xgb = modelo_xgb.predict_proba(X_test)[:, 1]

from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(
    y_test,
    y_prob_xgb
)

print("ROC-AUC:", roc_auc)
```

##### Conclusão - Modelo XGBoost

O modelo XGBoost foi treinado utilizando os conjuntos de treino e teste preparados na Sprint 4.

O modelo apresentou Accuracy de 86,05%, Precision de 65,00%, Recall de 27,66%, F1 Score de 38,81% e ROC-AUC de 76,38%.

Comparado aos modelos anteriores, o XGBoost apresentou o melhor equilíbrio entre identificação de desligamentos e taxa de acerto geral. O modelo foi capaz de identificar 13 dos 47 funcionários que efetivamente deixaram a empresa, superando os resultados obtidos pela Regressão Logística e pelo Random Forest.

O ROC-AUC de 0,764 indica boa capacidade de distinção entre funcionários com risco de desligamento e funcionários com tendência de permanência, tornando o XGBoost o modelo com melhor desempenho até o momento.
