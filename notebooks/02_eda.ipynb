### Sprint 4 – Preparação dos Dados
- Task 4.1 – Remoção de Variáveis Irrelevantes 
- Task 4.2 – Tratamento das Variáveis Categóricas 
- Task 4.3 – Preparação da Variável Alvo 
- Task 4.4 – Separação de Treino e Teste 

```python
# Verificação de colunas

df[
    [
        "EmployeeCount",
        "Over18",
        "StandardHours",
        "EmployeeNumber"
    ]
].describe(include="all")
```

```python
# Remoção das Variáveis

df = df.drop(
    columns=[
        "EmployeeCount",
        "Over18",
        "StandardHours",
        "EmployeeNumber"
    ]
)
```

```python
# Salvar o arquivo tratado

df.to_csv(
    DATA_PROCESSED_DIR / "employee_attrition_clean.csv",
    index=False
)
```

```python
# Verificar o que ainda é categórico

df.select_dtypes(include="object").columns.tolist()
```

```python
# Remover Attriction da Análise One-Hot Encoding

target = df["Attrition"]

features = df.drop(
    columns=["Attrition"]
)

features.select_dtypes(include="object").columns.tolist()
```

```python
# One-Hot Encoding

X = pd.get_dummies(
    features,
    drop_first=True,
    dtype=int
)
```

```python
X.select_dtypes(include="object").columns.tolist()
```

```python
X.head()
```

```python
# Target Encoding

y = df["Attrition"]
```

```python
# Target Encoding

y = target.map({
    "No": 0,
    "Yes": 1
})

y.head()
```

```python
# Importar função para separar treino e teste

from sklearn.model_selection import train_test_split
```

```python
# Separar os dados em treino e teste

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
# Validar a proporção de classes

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)

print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
# Validar balance das classes

print(y_train.value_counts(normalize=True) * 100)
print()
print(y_test.value_counts(normalize=True) * 100)
```

```python
# Exportar os dados de treino e teste

X_train.to_csv(
    DATA_PROCESSED_DIR / "X_train.csv",
    index=False
)

X_test.to_csv(
    DATA_PROCESSED_DIR / "X_test.csv",
    index=False
)

y_train.to_csv(
    DATA_PROCESSED_DIR / "y_train.csv",
    index=False
)

y_test.to_csv(
    DATA_PROCESSED_DIR / "y_test.csv",
    index=False
)
```
